## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    #your code here 
    for number in range(1, n + 1):
      if number == 0 or number == 1:
        rdo = 1
      else:
        rdo *= number
    return rdo

factorial(5)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... ðŸ§`

```python
lst = [4, 5, 5, 567, 8745, 56, 5, 3, 9]

def mode_counter(arr):
#your code here
  # Initialize a dictionary to map each number to its frequency 
  values_rep = {}
  
  for num in arr:    
    # Check if the number is already present in the dictionary
    if num in values_rep:
      # Increment the existing frequency coutn by one
      values_rep[num] += 1
    else:
      # If not found, initialize the number's frequency to one
      values_rep[num] = 1   

  # Find the maximum frequency value among all elements 
  max_times = max(values_rep.values())  

  the_mode = [key for key, value in values_rep.items() if value == max_times][0]
  
  # Return the mode
  return the_mode

mode_counter(lst)

  
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
#your code here

# Mean
def mean(lst_mean):
  # nÃºmero total de elementos
  total = len(lst_mean)
  sum_of_values = sum(lst_mean)
  return sum_of_values / total

def variance(lst_variance):
  # First step
  mean_cal = mean(lst_variance)
  lst_first_step = [(num - mean_cal)**2 for num in lst_variance]

  # Second step
  sum_of_squares = sum(lst_first_step)

  # Third step
  return sum_of_squares / (len(lst_variance) - 1)

def st_dev(list_sd):
  rdo = variance(list_sd)**0.5
  return rdo

print(st_dev(lst))
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ðŸ¤ª`

```python
def sort_alpha(string):
    # Reference alphabet to validate alphanumeric characters
    alpha = 'abcdefghijklmnopqrstuvwxyz' 
    # Normalize input to lowercase for consistent comparison
    string = string.lower()
    
    # Store the last index to handle the boundary condition at the end of the string
    last_index = len(string) - 1
    # Initialize a list to store identified words and a buffer for the current word
    list_of_words = []
    words = ''

    # 1. Iterate through the string character by character
    for index in range(len(string)):
        letter = string[index]
        # If the character is in the alphabet, append it to the current word 
        if letter in alpha:
            words += letter        
        # If a comma is found or we reach the end of the string, store the word
        if letter == ',' or index == last_index:
            if len(words) > 0:
              list_of_words.append(words)
              # Reset 'words' for the next word
              words = ''

    # 2. Sort the word list in alphabetical order
    list_of_words.sort()
    
    # 3. Serialize the list back into a string, manually handling delimiters
    words = ''
    num_words = len(list_of_words)
    
    # Iterate through the sorted list word by word
    for index in range(num_words):
        words += list_of_words[index]
        # Append a comma ONLY if it is not the last element to avoid trailing commas
        if index < num_words - 1:
            words += ','
            
    return words
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
  #your code here
  check_char = { 
    'lower_case': False,
    'upper_case': False,
    'number': False,
    'special_char': False
  }

  alpha_lower = 'abcdefghijklmnopqrstuvwxyz'
  alpha_upper = 'ABDCDEFGHIJKLMNOPQRSTUVWXYZ'
  number = '0123456789'
  # special_character = '!@#$%&*()-_+={[]}|;:,.<>?/~ '
  
  if len(password) >= 8:
    for character in password:
      if character in alpha_lower:
        check_char['lower_case'] = True
      elif character in alpha_upper:
        check_char['upper_case'] = True
      elif character in number:
        check_char['number'] = True
      else:
        check_char['special_char'] = True
  else:
    return False
  
  for value in check_char.values():
    if value == False:
      return False
  
  return True


check_pass('9@IronHack')
```
